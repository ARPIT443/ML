```
import cv2
from google.colab.patches import cv2_imshow
```

```
# reading image
img1 = cv2.imread("dog2.jpg")
```

```
# showing image 1
cv2_imshow(img1)       # use cv2.imshow() if not using google colab
```

```
img1.shape
```

```
# loading second image
img2 = cv2.imread("dog3.jpg")
```

```
# showing img2
cv2_imshow(img2)
```

```
img2.shape
```

# Adding two images

```
comb_img = cv2.add(img1,img2)
```

```
cv2_imshow(comb_img)
```

```
comb_img1 = cv2.addWeighted(img1, 0.5,img2, 0.5, 0) 
```

```
cv2_imshow(comb_img1)
```

```
import numpy as np
```

```
z = np.concatenate((img1,img2),axis=1)
```

```
cv2_imshow(z)
```

```
z1 = np.concatenate((img1,img2),axis=0)
```

```
cv2_imshow(z1)
```

# Find the difference between two images

```
img1.shape
```

```
img2.shape
```

```
diff = cv2.absdiff(img1,img2)
```

```
cv2_imshow(diff)
```
