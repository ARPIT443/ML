```
import pandas as pd
```

```
# loading data
df=pd.read_csv("social.csv")
```

```
# schema
df.info()
```

```
df.head()
```

```
# removing user id column
filter_data=df.iloc[:,2:].values
```

```
# now we can aply training test
features=filter_data[:,:2]
```

```
# label finding
label=filter_data[:,2]
```

```
# training and testing
from sklearn.model_selection import train_test_split
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```
# importing function for feature scaling
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
```

```
# now apply feature scalling in training data only
X=sc.fit_transform(X)
```

```
X
```

```
# training data scalling
x=sc.transform(x)
```

```
# KNN and Random Forest
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```
# random forest call with number of decision tree
Rclf=RandomForestClassifier(n_estimators=20)
```

```
# training model
trained=Rclf.fit(X,Y)
```

```
# now predicting 
output=trained.predict(x)
```
