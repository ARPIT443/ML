```
import cv2
from google.colab.patches import cv2_imshow
```

```
# loading image
img = cv2.imread("dog.jpg")
```

```
#showing image
# cv2.imshow(img)  ----> it does not support in google colab
cv2_imshow(img)
```

```
# checking shape of image
img.shape
```

```
# resizing of image
resize_img = cv2.resize(img,(150,150))
```

```
cv2_imshow(resize_img)
```

```
# checking new img shape
resize_img.shape
```

#***Compress***

```
img.shape    # original shape
```

```
cv2.imwrite('compress_img.jpg',img,[cv2.IMWRITE_JPEG_QUALITY,0])
```

```
# reading compressed image
com_img = cv2.imread("compress_img.jpg")
```

```
# showing img
cv2_imshow(com_img)
```

```
com_img.shape
```

# Decompress

```
cv2.imwrite('Decompress_img.jpg',img,[cv2.IMWRITE_JPEG_QUALITY,90])
```

```
# loading decompress image
decom_img = cv2.imread("Decompress_img.jpg")
```

```
cv2_imshow(decom_img)
```

```
decom_img.shape
```
